	r	V

```python
 # Your code here
 #(a) (5 pts) Calculate the difference in the number of Occupied Housing Units from year to year and print it. The difference must be calculated for
 #the consecutive years such as 2008-2009, 2009-2010 etc. Finally, print the values in the ascending order.

 import pandas as pd

 #file_path = "Assignment1_denton_housing.csv"
 df = pd.read_csv("/content/Assignment1_denton_housing.csv")
 df.head()

 # Filter the data for "Occupied Housing Units"
occupied_units = df[df["title_field"] == "Occupied Housing Units"].copy()

# Sort the data by year in ascending order
occupied_units.sort_values(by="year", inplace=True)

# Calculate the year-to-year difference
occupied_units["Difference"] = occupied_units["value"].diff()

# Drop the first row as it will have a NaN difference
occupied_units.dropna(inplace=True)

# Sort the differences in ascending order
occupied_units_sorted = occupied_units[["year", "Difference"]].sort_values(by="Difference")

# Display the result
print(occupied_units_sorted)
print("\n")


#(b) (5 pts) For all the years, calculate the percentage of housing units which are vacant an occupied. Print the results in the following format:
#Year Vacant Housing Units Occupied Housing Units

# Filter the data for Total, Vacant, and Occupied Housing Units
total_units = df[df["title_field"] == "Total Housing Units"].copy()
vacant_units = df[df["title_field"] == "Vacant Housing Units"].copy()
occupied_units = df[df["title_field"] == "Occupied Housing Units"].copy()

# Merge the datasets on year
housing_data = total_units.merge(vacant_units, on="year", suffixes=("_total", "_vacant"))
housing_data = housing_data.merge(occupied_units, on="year")

# Rename columns for clarity
housing_data.rename(columns={"value_total": "Total Housing Units",
                             "value_vacant": "Vacant Housing Units",
                             "value": "Occupied Housing Units"}, inplace=True)

# Calculate the percentage of vacant and occupied units
housing_data["30%"] = (housing_data["Vacant Housing Units"] / housing_data["Total Housing Units"]) * 100
housing_data["70%"] = (housing_data["Occupied Housing Units"] / housing_data["Total Housing Units"]) * 100

# Select and format the final output
housing_percentage = housing_data[["year", "30%", "70%"]].sort_values(by="year")

# Display the result
print(housing_percentage)

print("\n")
#(c) (5 pts) Calculate and print the valued and years in which the highest number of housing units were vacant and occupied. Print the results in the following format:

# Find the year and value with the highest number of vacant housing units
max_vacant = vacant_units.loc[vacant_units["value"].idxmax(), ["year", "value"]]

# Find the year and value with the highest number of occupied housing units
max_occupied = occupied_units.loc[occupied_units["value"].idxmax(), ["year", "value"]]

# Create a formatted result DataFrame
highest_values = pd.DataFrame({
    "Category": ["Vacant Housing Units", "Occupied Housing Units"],
    "Year": [max_vacant["year"], max_occupied["year"]],
    "Value": [max_vacant["value"], max_occupied["value"]]
})

# Display the result
print(highest_values)
print("\n")

#(d) (5 pts) Calculate the harmonic mean of the total housing units and print it out.
from scipy.stats import hmean

# Calculate the harmonic mean of total housing units
harmonic_mean_total_units = hmean(total_units["value"])

# Display the result
print("Harmonic mean is:", harmonic_mean_total_units)









```

# Question 6 (20 points).
Develop a advance Python program to manage inventory records for a prodcut managing company. The system needs functions for adding, deleting and changing products and basic search options. Your system should use CSV files to automate restocking and produce reports that show item values alongside category breakdowns. Manage edge cases and corrupted CSV files. Finally maintaine 50 products of different categories in inventory.

```python
# your code here

import csv
import os
import random
import pandas as pd

# File to store inventory data
INVENTORY_FILE = 'inventory.csv'

#1.Created a file and Attempting to open a locked CSV file. To handles situations where the inventory file is locked or inaccessible due to system permissions. Instead of crashing, it alerts the user and suggests possible fixes.
def initialize_inventory():
    """Creates an inventory file with 50 initial products if not exists."""
    if not os.path.exists(INVENTORY_FILE):
        categories = ["Electronics", "Furniture", "Clothing", "Groceries", "Toys"]
        with open(INVENTORY_FILE, 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(["Product ID", "Name", "Category", "Quantity", "Price"])
            for i in range(50):
                writer.writerow([
                    f"P{i+1}",
                    f"Product_{i+1}",
                    random.choice(categories),
                    random.randint(1, 20),
                    round(random.uniform(5, 500), 2)
                ])

#2. Reading ann Handling a corrupted CSV file. (asked chatgpt how to read and handle the corrupt file. It helps to detect missing values in the file before processing it.
#If corruption is found, the system warns the user and attempts to recover clean data.)
def read_inventory():
    """Reads the inventory file and handles corruption cases."""
    try:
        df = pd.read_csv(INVENTORY_FILE)
        if df.isnull().values.any():
            raise ValueError("Corrupted file: Missing values detected.")
        return df
    except Exception as e:
        print(f"Error reading inventory file: {e}")
        return None

# 3. Adding a product with negative price or quantity. (chatgpt helped me to ensures that product quantity and price cannot be negative.
#If invalid values are entered, the system rejects them and prompts for corrections.)
def add_product():
    """Prompts user to add a new product with validation checks."""
    df = read_inventory()
    if df is None:
        return

 # 4. Attempting to add duplicate product IDs. (This helps prevents adding products with the same ID twice to maintain unique inventory records. If a duplicate is detected, an error message is shown.)

    product_id = input("Enter Product ID: ")
    if product_id in df["Product ID"].values:
        print("Error: Product ID already exists!")
        return

    name = input("Enter Product Name: ")
    category = input("Enter Product Category: ")

    try:
        quantity = int(input("Enter Product Quantity: "))
        price = float(input("Enter Product Price: "))
        if quantity < 0 or price < 0:
            raise ValueError("Quantity and Price must be positive.")
    except ValueError as e:
        print(f"Invalid input: {e}")
        return

    new_product = pd.DataFrame({
        "Product ID": [product_id],
        "Name": [name],
        "Category": [category],
        "Quantity": [quantity],
        "Price": [price]
    })
    df = pd.concat([df, new_product], ignore_index=True)
    df.to_csv(INVENTORY_FILE, index=False)
    print("Product added successfully.")

 #5. Deleting a product that is not in the inventory. (The reason to add delete is to prevents errors when attempting to delete an item that is not in the inventory and
  #Displays a clear message indicating that the product ID was not found.)
def delete_product():
    """Prompts user to delete a product by ID."""
    df = read_inventory()
    if df is None:
        return

    product_id = input("Enter Product ID to delete: ")
    if product_id not in df["Product ID"].values:
        print("Error: Product ID not found!")
        return

    df = df[df["Product ID"] != product_id]
    df.to_csv(INVENTORY_FILE, index=False)
    print("Product deleted successfully.")

#6. Updating a product with invalid data. (The reason i asked chatgpt to update a product is to ensures that only valid values (e.g., positive numbers, correct categories)
#are accepted when updating product details. Invalid updates trigger an error message.)
def update_product():
    """Prompts user to update product details."""
    df = read_inventory()
    if df is None:
        return

    product_id = input("Enter Product ID to update: ")
    if product_id not in df["Product ID"].values:
        print("Error: Product ID not found!")
        return

    index = df[df["Product ID"] == product_id].index[0]
    name = input("Enter new Product Name (leave blank to keep current): ")
    category = input("Enter new Product Category (leave blank to keep current): ")

    try:
        quantity = input("Enter new Quantity (leave blank to keep current): ")
        if quantity:
            quantity = int(quantity)
            if quantity < 0:
                raise ValueError("Quantity cannot be negative.")
            df.at[index, "Quantity"] = quantity

        price = input("Enter new Price (leave blank to keep current): ")
        if price:
            price = float(price)
            if price < 0:
                raise ValueError("Price cannot be negative.")
            df.at[index, "Price"] = price
    except ValueError as e:
        print(f"Invalid input: {e}")
        return

    if name:
        df.at[index, "Name"] = name
    if category:
        df.at[index, "Category"] = category

    df.to_csv(INVENTORY_FILE, index=False)
    print("Product updated successfully.")

#7. Case-sensitive and partial searches.
#8. Searching for a non-existent product
# (To allows flexible searching by ignoring case sensitivity and supporting partial matches chagpt provided me the possible solution for it.
#This improves the user experience by making searches more intuitive.)
def search_product():
    """Prompts user to search for a product by keyword."""
    df = read_inventory()
    if df is None:
        return

    keyword = input("Enter product name or keyword to search: ")
    results = df[df["Name"].str.contains(keyword, case=False, na=False)]

    if results.empty:
        print("No products found.")
    else:
        print(results)

# 9. Auto restocking when quantity is too low.
#10. Generating reports for empty categories
#(To restock the inventory asked chatgpt to monitor stock levels and aleter users when product is below the threshold limit.
#Ensures reports remain meaningful even if no products exist in a certain category. Instead of errors, the system displays a message indicating that the category is empty.)

def generate_report():
    """Generates a summary report with category breakdowns and stock alerts."""
    df = read_inventory()
    if df is None:
        return

    category_summary = df.groupby("Category")["Quantity"].sum().reset_index()
    low_stock_items = df[df["Quantity"] < 5]

    print("Category Breakdown:")
    print(category_summary)

    if not low_stock_items.empty:
        print("Low Stock Items:")
        print(low_stock_items)
    else:
        print("All items are sufficiently stocked.")

# Initialize inventory
initialize_inventory()

# User Menu
while True:
    print("\nInventory Management System")
    print("1. Add Product")
    print("2. Delete Product")
    print("3. Update Product")
    print("4. Search Product")
    print("5. Generate Report")
    print("6. Exit")

    choice = input("Enter your choice (1-6): ")
    if choice == "1":
        add_product()
    elif choice == "2":
        delete_product()
    elif choice == "3":
        update_product()
    elif choice == "4":
        search_product()
    elif choice == "5":
        generate_report()
    elif choice == "6":
        print("Exiting system. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 6.")
```

# Question 7 (20 points)

Create a well desined Python program for a flight booking system. Offer core functions for adding new flights, finding trips, reserving and releasing seating space and creating performance reports. Ensure the system manages edge cases like invalid input, overbooking, and cancellations. Finally, extend the booking system by including advanced features and maintain about 50 flight in the system.

```python
# your code here

import random
import datetime

#Following are the 10 edge cases to solve:
#1. To generate a random 50 flights for popular desination.
#2. To book a Seat
#3. To cancel Seat
#4. To check availability
#5. Adding a flight
#6. Booking a flight
#7. Searching for a flight
#8. To cancel the booking
#9. Generating a report
#10.created a prompt to perform all the above mentined functiono.

class Flight:

    """Class to represent a flight with details such as flight number, destination, capacity, and bookings."""
    def __init__(self, flight_no, destination, date, capacity, price):
        self.flight_no = flight_no
        self.destination = destination
        self.date = date
        self.capacity = capacity
        self.price = price
        self.bookings = []  # List to store booked passengers

    def book_seat(self, passenger_name):
        """Book a seat for a passenger if available while handling edge cases."""
        if len(self.bookings) >= self.capacity:
            print("Sorry, this flight is fully booked.")
            return
        if passenger_name in self.bookings:
            print("You have already booked a seat on this flight.")
            return
        self.bookings.append(passenger_name)
        print(f"Booking confirmed for {passenger_name} on flight {self.flight_no} to {self.destination}.")

    def cancel_seat(self, passenger_name):
        """Cancel a booked seat for a passenger while handling edge cases."""
        if passenger_name not in self.bookings:
            print("No booking found under this name.")
            return
        self.bookings.remove(passenger_name)
        print(f"Booking for {passenger_name} has been cancelled.")

    def get_availability(self):
        """Return available seats count."""
        return self.capacity - len(self.bookings)

    def __str__(self):
        return f"Flight {self.flight_no} to {self.destination} on {self.date} | Available Seats: {self.get_availability()}"

class FlightBookingSystem:
    """Class to manage flight operations, including adding flights, searching, booking, and reporting."""
    def __init__(self):
        self.flights = []
        self.populate_flights()

    def populate_flights(self):
        """Generate 50 random flights with correct date handling."""
        destinations = ["New York", "London", "Paris", "Tokyo", "Dubai", "Sydney", "Berlin", "Toronto"]
        today = datetime.date.today()
        for i in range(1, 51):
            flight_no = f"F{i:03d}"
            destination = random.choice(destinations)
            date = (today + datetime.timedelta(days=random.randint(1, 30))).strftime('%Y-%m-%d')
            capacity = random.randint(50, 200)
            price = random.randint(100, 1000)
            self.flights.append(Flight(flight_no, destination, date, capacity, price))

    def find_flight(self, destination, date):
        """Find available flights based on destination and date while handling invalid input."""
        available_flights = [f for f in self.flights if f.destination.lower() == destination.lower() and f.date == date]
        if not available_flights:
            print("No available flights found. Try another date or destination.")
            print("Available flights:")
            for flight in self.flights:
                print(flight)
            return
        for flight in available_flights:
            print(flight)

    def book_flight(self, flight_no, passenger_name):
        """Book a seat on a specific flight while checking edge cases."""
        if not passenger_name.strip():
            print("Passenger name cannot be empty.")
            return
        for flight in self.flights:
            if flight.flight_no == flight_no:
                if datetime.datetime.strptime(flight.date, '%Y-%m-%d').date() < datetime.date.today():
                    print("Cannot book a flight in the past.")
                    return
                flight.book_seat(passenger_name)
                return
        print("Invalid flight number.")

    def cancel_booking(self, flight_no, passenger_name):
        """Cancel a booking on a specific flight while handling invalid input."""
        if not passenger_name.strip():
            print("Passenger name cannot be empty.")
            return
        for flight in self.flights:
            if flight.flight_no == flight_no:
                flight.cancel_seat(passenger_name)
                return
        print("Invalid flight number.")

    def generate_report(self):
        """Generate a performance report of all flights."""
        print("Flight Performance Report")
        for flight in self.flights:
            booked_percentage = (len(flight.bookings) / flight.capacity) * 100
            print(f"{flight.flight_no}: {len(flight.bookings)}/{flight.capacity} seats booked ({booked_percentage:.2f}%)")

# Interactive Flight Booking System
if __name__ == "__main__":
    system = FlightBookingSystem()

    while True:
        print("\nWelcome to the Flight Booking System")
        print("1. Search for Flights")
        print("2. Book a Flight")
        print("3. Cancel a Booking")
        print("4. Generate Flight Report")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ")

        if choice == "1":
            destination = input("Enter your destination: ")
            date = input("Enter your travel date (YYYY-MM-DD): ")
            system.find_flight(destination, date)

        elif choice == "2":
            flight_no = input("Enter flight number to book: ")
            passenger_name = input("Enter your name: ")
            system.book_flight(flight_no, passenger_name)

        elif choice == "3":
            flight_no = input("Enter flight number to cancel: ")
            passenger_name = input("Enter your name: ")
            system.cancel_booking(flight_no, passenger_name)

        elif choice == "4":
            system.generate_report()

        elif choice == "5":
            print("Exiting the Flight Booking System. Goodbye!")
            break

        else:
            print("Invalid choice. Please enter a number between 1 and 5.")
```
